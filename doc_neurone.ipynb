## 1. Qu’est-ce qu’un Perceptron ?

Le **perceptron** est le premier modèle de **réseau de neurones artificiels**, proposé par **Frank Rosenblatt (1957)**.
Il s’agit d’un **classificateur linéaire binaire**.

### Lien entre neurone biologique et perceptron

| Neurone biologique | Perceptron     |
| ------------------ | -------------- |
| Dendrites          | Entrées xᵢ     |
| Synapses           | Poids wᵢ       |
| Corps cellulaire   | Somme pondérée |
| Seuil d’activation | Biais b        |
| Axone              | Sortie y       |

---

## 2. Fonction mathématique du Perceptron

### Équation du perceptron

$$
y = f\left(\sum_{i=1}^{n} w_i x_i + b\right)
$$

### Définition des termes

* xᵢ : entrées (caractéristiques)
* wᵢ : poids associés aux entrées
* b : biais
* Σ(wᵢxᵢ) + b : somme pondérée
* f(.) : fonction d’activation
* y : sortie du perceptron

### Usage

* Classification binaire
* Séparation linéaire
* Détection de motifs simples

---

## 3. Règle d’apprentissage du Perceptron

Pour chaque exemple (x, y) :

$$
w_i = w_i + \eta (y - \hat{y}) x_i
$$

$$
b = b + \eta (y - \hat{y})
$$

Où :

* η : taux d’apprentissage
* y : sortie attendue
* ŷ : sortie prédite

---

## 4. Fonction d’activation

Le perceptron utilise une **fonction seuil** :

$$
f(z) =
\begin{aligned}
1 \quad \text{si } z \ge 0 \
0 \quad \text{si } z < 0
\end{aligned}
$$


---

## 5. Processus d’entraînement

1. Initialisation des poids et du biais
2. Présentation d’un exemple
3. Calcul de la sortie
4. Comparaison avec la sortie attendue
5. Mise à jour en cas d’erreur
6. Répétition sur plusieurs époques

### Convergence

Le perceptron converge **si et seulement si** les données sont **linéairement séparables**.

---

## 6. Limites du Perceptron

* Problèmes non linéaires impossibles à résoudre
* Échec sur le problème XOR
* Modèle à une seule couche
* Fonction non dérivable
* Sensible au bruit
